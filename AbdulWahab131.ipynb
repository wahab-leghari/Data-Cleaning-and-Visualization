```python
data=[13,15,16,16,19,20,20,21,22,22,25,25,25,25,30,33,33,35,35,35,35,36,40,45,46,52,70]
```

# a)Mean And Median

```python
mean=np.mean(data)
median=np.median(data)
print("Mean= ", mean)
print("Median= ",median)
```

# b) Mode And Modality

```python
mode=stats.mode(data)
print("Mode= ",mode)
```

```python
df=pd.DataFrame(data)
mode=df.mode()
print("Mode:",mode)
if mode.count()>1:
    if mode.count()>2:
        if mode.count()>3:
            print("Data is Multi-modal")
        else:
            print("Data is Tri-modal")
    else:
        print("Data is Bi-modal")
else:
    print("Data is Uni-modal")
```

# c)Mid Range

```python
Mid_range= (min(data)+max(data))/2 
print("Mid Range= ",Mid_range)
```

# d)First and Third Quartile

# Part 2 Using forest fire dataset

```python
forestFire=pd.read_csv('forestfires.csv')
forestFire.dtypes
```

Box PLot

```python
box_plot=[forestFire['FFMC'],forestFire['DMC'],forestFire['DC'],forestFire['ISI'],forestFire['temp'],forestFire['RH'],forestFire['wind'],forestFire['rain'],forestFire['area']]
plt.boxplot(box_plot,patch_artist=True,labels=['FFMC','DMC','DC','ISI','temp','RH','wind','rain','area'])
```

Mode

```python
print("Mode of Month= ",stats.mode(forestFire['month']))
print("Mode of day= ",stats.mode(forestFire['day']))
```

QQ plot of keys area and temp

```python
sm.qqplot(data=forestFire['area'],fit=True,line='q',dist='norm')
plt.title('Q-Q plot for month ',size=17)
plt.show()
sm.qqplot(data=forestFire['temp'],fit=True,line='q',dist='norm')
plt.title('Q-Q plot for temp ',size=17)
plt.show()
```

Histogram 

```python
plt.hist([forestFire['area'],forestFire['temp']], bins = 10)
plt.title('Histogram of area and temp ',size=17)
```

Scattar Plot

```python
plt.scatter(forestFire['area'],forestFire['temp'],c='b',marker='H')
plt.xlabel('temp', fontsize=14)
plt.ylabel('area', fontsize=14)
plt.title('scatter plot - temp vs area',fontsize=20)
plt.show()
```

```python
corr = forestFire.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```
